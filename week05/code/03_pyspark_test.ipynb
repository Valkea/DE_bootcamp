```python
from pathlib import Path

import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('test') \
    .getOrCreate()
```

```python
df = spark.read \
    .option("header", "true") \
    .csv(str(Path('data','taxi+_zone_lookup.csv')))
```

```python
df.show()
```

```python
# Write parquet file to 'zones' folder
df.write.parquet('data/zones')
```
